```python
import numpy as np
from scipy.optimize import curve_fit

# Definindo a equação exponencial
def exponential_func(x, A, B, C):
    return A * np.exp(B * x) + C

# Dados de exemplo
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2.1, 3.9, 7.2, 10.8, 17.2])

# Realizando o ajuste de regressão
popt, pcov = curve_fit(exponential_func, x_data, y_data)

# Obtendo os valores estimados para A, B e C
A_est, B_est, C_est = popt

# Obtendo as incertezas associadas
A_err, B_err, C_err = np.sqrt(np.diag(pcov))

# Calculando os erros relativos
A_rel_err = A_err / A_est
B_rel_err = B_err / B_est
C_rel_err = C_err / C_est

# Exibindo os resultados
print("Valores estimados:")
print("A:", A_est)
print("B:", B_est)
print("C:", C_est)
print("\nIncertezas associadas:")
print("Incerteza em A:", A_err)
print("Incerteza em B:", B_err)
print("Incerteza em C:", C_err)
print("\nErros relativos:")
print("Erro relativo em A:", A_rel_err)
print("Erro relativo em B:", B_rel_err)
print("Erro relativo em C:", C_rel_err)
```
